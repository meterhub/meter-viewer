```python
from matplotlib import pyplot as plt
import random
```

```python
nums = [random.randint(0, 99999) for i in range(100000)]
```

```python
data = nums.copy()
```

```python
plt.hist(data, bins=30, color='skyblue', alpha=0.8)
```

```python
num_count = {}
def increase(i: int):
    num_count[i] = num_count.get(i, 0) + 1
```

```python
for i in nums:
    si = str(i)
    si_list = list(si)
    increase(int(si_list[-1]))
```
