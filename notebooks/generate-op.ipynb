```python
from meterviewer.datasets import dataset, single
from meterviewer.config import get_root_path
from meterviewer.img import process
```

```python
root_path = get_root_path()

def gen(digit_list = ["1", "2", "3", "4"],):
    im = dataset.generate_block_img(
        digit_list,
        process.join_img,
        single.read_rand_img,
    )
    process.show_img(im, is_stop=0)
```

```python
gen(list('12399'))
```

```python
dataset_name = "M10L8XL"
```

```python
root_path = get_root_path()

def gen(digit_list = ["1", "2", "3", "4"],):
    im = dataset.generate_block_img(
        digit_list,
        root_path,
        dataset.join_with_resize,
        lambda: dataset_name,
        single.read_rand_img,
    )
    img.show_img(im, is_stop=0)
```
