```python
from meterviewer.datasets import dataset, single
from meterviewer.labeling.config import get_root_path
from meterviewer import img
```

```python
root_path = get_root_path()

def gen(digit_list = ["1", "2", "3", "4"],):
    im = dataset.generate_block_img(
        digit_list,
        root_path,
        img.join_img,
        lambda: "M1L1XL",
        single.read_rand_img,
    )
    img.show_img(im, is_stop=0)
```

```python
gen(list('12399'))
```
